```python
import mmpose
from mmpose.apis import MMPoseInferencer
import cv2
import numpy as np



video_path = "./video/Kamera1.dav "
video_capture = cv2.VideoCapture(video_path)

#get video size
w, h, = (

    int(video_capture.get(x))

    for x in (

        cv2.CAP_PROP_FRAME_WIDTH,

        cv2.CAP_PROP_FRAME_HEIGHT,

    )

)
#create result file
result = cv2.VideoWriter("out_video.avi", cv2.VideoWriter_fourcc(*"mp4v"), 20, (w, h)) #fps, (2 * w, 2 * h) mp4v MJPG

 



if not video_capture.isOpened():

    raise ValueError(f"Could not open video file: {video_path}")

#skip frames at start where there is no people to detect (fps * time to skip[s])
skip_frames = 200

for _ in range(skip_frames):

      ret, frame = video_capture.read()


inferencer = MMPoseInferencer('human')

ret = True

# licz = 0

while ret:
    ret, frame = video_capture.read()
 
    if not ret:
         break


    
    result_generator = inferencer(frame, show=False,return_vis = True)  # turn off showing the result for speed
    result_frame = next(result_generator)
    xx = result_frame['visualization']
    xx = np.array(xx[0])
    
    xx = cv2.cvtColor(xx, cv2.COLOR_BGR2RGB)
    
   
    
    print('Yes')
    

    result.write(xx)
    # licz += 1

video_capture.release()
result.release()
```
